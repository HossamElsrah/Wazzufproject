```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import os
```

```python
companies_lst_ml,titles_lst_ml, links_lst_ml, occupations_lst_ml=  [] , [] , [] , []
for i in range(0,6):
  response = requests.get(f'https://wuzzuf.net/search/jobs/?a=hpb&q=machine%20learning&start={i}')
  soup = BeautifulSoup(response.content , 'lxml')
  titles = soup.find_all("h2", {'class': 'css-m604qf'})
  titles_lst_ml += [title.a.text for title in titles]
  links_lst_ml += ['https://wuzzuf.net' + title.a['href'].replace(' ', '%20') for title in titles]
  occupations = soup.find_all("div", {'class': 'css-1lh32fc'})
  occupations_lst_ml += [occupation.text for occupation in occupations]
  Company_Location = soup.find_all("span" , {'class' : 'css-5wys0k'})
  Company_Location_lst =[ loc.text.replace('t ' , 't') for loc in Company_Location ]
  companies_lst_ml +=  Company_Location_lst

print(titles_lst_ml,links_lst_ml,occupations_lst_ml,companies_lst_ml)
```
